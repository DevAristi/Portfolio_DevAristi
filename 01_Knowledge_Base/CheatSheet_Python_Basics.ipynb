# Module 1: Python Basics (Cheat Sheet)
**FOCUS:** Basic Syntax, Data Types, and String Manipulation.
**NOTE:** Lines starting with '#' are comments ignored by Python.

```python
# --- 1. VARIABLES & ASSIGNMENT ---
# Syntax: variable_name = value

# Integers - Whole numbers without decimals
x = 7
age = 19

# Floats - Numbers with decimals
y = 12.4
price = 99.99

# Booleans - True or False (Critical for logic)
is_valid = True
has_access = False

# Strings - Text always inside quotes
user_name = "DevAristi" # Just a variable name.

# Strings - Text always inside quotes
# The 'f' before the quotes turns the string into a "template".
# It allows us to put variables INSIDE the curly brackets {}.
print(f"User: {user_name}, Age: {age}")
# Without 'f', it would print: "User: {user_name}, Age: {age}" (This looks horrible).
```

```python
# --- 2. MATH OPERATORS ---

a = 10
b = 3

# Basic Operations
addition = a + b            #13
substraction = a - b        #7
multiplication = a * b      #30

# --- KEY DIFFERENCES ---

# Standard Division (/) - Always returns a float.
division = a / b        #3.3333...

# Floor Division (//) Cuts off decimals, returns int.
floor_div = a // b      #3 (Great for Index Math)

# Modulo (%) - Returns the REMAINDER
# Trick: If number % 2 == 0, it's EVEN. If 1, it's ODD.
remainder = a % b   # 1 (3 fits into 10 three times, remainder 1)

print(f"Division: {division}")
print(f"Floor Div: {floor_div}")
print(f"Modulo: {remainder}")
```

```python
# --- 3. STRING MANIPULATION ---
text = "    Hello, Python World  "

# Cleaning (Strip) - Removes leading/trailing whitespace
clean_text = text.strip()
print(f"Cleaned: '{clean_text}") # "Hello, Python World"

# Transformation (Upper/Lower) - Useful for data normalization
print(clean_text.upper())   # "HELLO, PYTHON WORLD"
print(clean_text.lower())   # "hello, python world"

# Replacement (Replace) - Swap parts of the text
# Syntax: .replace("old", "new")
new_text = clean_text.replace("World", "Engineer")
print(new_text)     # "Hello, Python Engineer"

# Splitting (Split) - Converts String ---> a List
# Crucial for passing CSV files or comm-separated data
data_list = clean_text.split(",")
print(data_list)            #['Hello', 'Python World']
```

```python
# --- 4. Indexing & Slicing ---
# Remember: Python uses 0-based indexing

string_data = "Python"

# Indexing (Access single character)
first_char = string_data[0]     #'P'
last_char = string_data[-1]     #'n' (Trick: -1 is always the last one)

# Slicing (Extract a substring)
# Syntax: [start : end] (The 'end' index is EXCLUSIVE/Not included)
slice_data = string_data[0:4] # "Pyth" (Indices 0, 1, 2, 3)

# Length (Len)
total_len = len(string_data)    #6

print(f"First: {first_char}, Last: {last_char}")
print(f"Slice: {slice_data}")
print(f"Total Length: {total_len}")
```
